# Explore Data, Create Model and Predict Data

The Iris dataset is one of the most famous datasets in the field of machine learning. It's often used for demonstrating various techniques in data visualization and machine learning. The dataset contains measurements of 150 iris flowers from three different species: setosa, versicolor, and virginica. The four measurements (features) are:

1. Sepal length: Sepals are the parts that encase and protect the flower bud before it blooms. The sepal length is the measurement from the base to the tip of the sepal.

2. Sepal width: This is the measurement of the width of the sepal at its broadest point.

3. Petal length: Petals are the parts that give a flower its shape and color. They are often the most noticeable part of a flower. Petal length is the measurement from the base to the tip of the petal.

4. Petal width: This is the measurement of the width of the petal at its broadest point.

All these measurements are taken in centimeters.

สร้าง Model ทำนาย

```python
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import load_iris

# Load the iris dataset
iris = load_iris()

# Standardize the features to have mean=0 and variance=1
scaler = StandardScaler()
X = scaler.fit_transform(iris.data)

# Split the data into training set and test set
X_train, X_test, y_train, y_test = train_test_split(X, iris.target, test_size=0.2, random_state=42)

# Initialize MLP classifier
mlp = MLPClassifier(hidden_layer_sizes=(10, 10, 10), max_iter=1000)

# Fit the data to the model
mlp.fit(X_train, y_train)

# Use the trained network to predict the test set
predictions = mlp.predict(X_test)

# Print the prediction results
print(predictions)
```

# Image Classification

## Load Images from dataset cifar10

```python
from tensorflow.keras import datasets
import matplotlib.pyplot as plt

# Load the CIFAR-10 dataset
(train_images, train_labels), (test_images, test_labels) = datasets.cifar10.load_data()

# Normalize pixel values to be between 0 and 1
train_images, test_images = train_images / 255.0, test_images / 255.0
```

## Display images

the labels in CIFAR-10 are coded as integers from 0 to 9. Each integer corresponds to a class of images as follows:

*   0: airplane
*   1: automobile
*   2: bird
*   3: cat
*   4: deer
*   5: dog
*   6: frog
*   7: horse
*   8: ship
*   9: truck

```python
class_names = ['airplane', 'automobile', 'bird', 'cat', 'deer', 'dog', 'frog', 'horse', 'ship', 'truck']
```

Display images

```python
plt.figure(figsize=(10,8))
for i in range(20):
    plt.subplot(5,4,i+1)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    plt.imshow(train_images[i], cmap=plt.cm.binary)
    # The CIFAR labels happen to be arrays,
    # which is why you need the extra index
    plt.xlabel(class_names[train_labels[i][0]])
plt.tight_layout()
plt.show()
```

## Train model

!! ก่อนรันอย่าลืมเปลี่ยน Runtime Type ให้เป็น GPU ก่อน

```python
import tensorflow as tf
from tensorflow.keras import layers, models

# Create the convolutional base
model = models.Sequential()
model.add(layers.Conv2D(32, (3, 3), activation='relu', input_shape=(32, 32, 3)))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(64, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(64, (3, 3), activation='relu'))

# Add Dense layers on top
model.add(layers.Flatten())
model.add(layers.Dense(64, activation='relu'))
model.add(layers.Dense(10))

# Compile and train the model
model.compile(optimizer='adam',
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])

history = model.fit(train_images, train_labels, epochs=10,
                    validation_data=(test_images, test_labels))

# Evaluate the model
test_loss, test_acc = model.evaluate(test_images,  test_labels, verbose=2)
print('\nTest accuracy:', test_acc)
```

## Use model to classify new image

```python
import numpy as np
from PIL import Image

# Load an image file to test, resizing it to 32x32 pixels (as required by this model)
img = Image.open('blue.jpg')


# Display the image
plt.imshow(img)
plt.show()

# Resize
img = img.resize((32, 32))

# Convert the image to a numpy array
img_array = np.array(img)

# Scale the image pixels by 255 (or use a scaler from sklearn here)
img_array = img_array / 255

# Insert a new axis to match the model's input
img_np = np.expand_dims(img_array, axis=0)

# Predict the class of the image
predictions = model.predict(img_np)

# The prediction for the image is the index of the class with the highest probability
class_index = np.argmax(predictions[0])

# Print the class index
print("Class index:", class_index)

# Convert the class index to the corresponding class name
class_name = class_names[class_index]
print("Class name:", class_name)
```

# Text Classification

The IMDB (Internet Movie Database) dataset is a set of 50,000 movie reviews for binary sentiment classification, provided as a part of the Keras dataset. There are 25,000 reviews for training and 25,000 reviews for testing. Each set includes an equal number of positive and negative reviews (i.e., 12,500 positive reviews and 12,500 negative reviews in each set).


In the IMDB dataset, the classes 1 and 0 represent the sentiment or sentiment polarity of the movie reviews.

* Class 1: Positive sentiment
* Class 0: Negative sentiment

Each review in the dataset is labeled as either positive or negative based on the sentiment expressed in the review text. The goal of sentiment analysis on this dataset is to classify the reviews correctly into one of these two classes.

For example, if a review has a positive sentiment expressing satisfaction or praise about a movie, it will be labeled as class 1. On the other hand, if a review has a negative sentiment expressing dissatisfaction or criticism about a movie, it will be labeled as class 0.

The binary classification task in the IMDB dataset is to build a model that can predict the sentiment of movie reviews accurately based on the provided training data.


The dataset is preprocessed, and each review (sequence of words) has been converted into a sequence of integers, where each integer represents a specific word in a dictionary. The dictionary can be retrieved with imdb.get_word_index().

```python
from keras.datasets import imdb

# Load the dataset
(X_train, y_train), (X_test, y_test) = imdb.load_data()

# Get the word index
word_index = imdb.get_word_index()

# Reverse the word index to get a dict where the keys are index numbers and the values are the corresponding words
index_word = {i: word for word, i in word_index.items()}

# Use the index_word dict to convert the first review into text
print(' '.join([index_word.get(i, '?') for i in X_train[0]]))
```

```python
from keras.datasets import imdb
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Load the dataset
(X_train, y_train), (X_test, y_test) = imdb.load_data()

# Get the word index
word_index = imdb.get_word_index()

# Reverse the word index to get a dict where the keys are index numbers and the values are the corresponding words
index_word = {i: word for word, i in word_index.items()}

# Get the words for the first 1000 reviews
words = ' '.join([' '.join([index_word.get(i - 3, '?') for i in review]) for review in X_train[:1000]])

# Create a word cloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(words)

# Display the word cloud
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
from tensorflow.keras.datasets import imdb
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM, Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences

# Load the dataset, but only keep the top n words, zero the rest
top_words = 5000
(X_train, y_train), (X_test, y_test) = imdb.load_data(num_words=top_words)



# Truncate and pad input sequences
max_review_length = 500
X_train = pad_sequences(X_train, maxlen=max_review_length)
X_test = pad_sequences(X_test, maxlen=max_review_length)

# Create the model
embedding_vector_length = 32
model = Sequential()
model.add(Embedding(top_words, embedding_vector_length, input_length=max_review_length))
model.add(LSTM(100))
model.add(Dense(1, activation='sigmoid'))
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# Train the model
model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=3, batch_size=64)

# Evaluate the model
scores = model.evaluate(X_test, y_test, verbose=0)
print("Accuracy: %.2f%%" % (scores[1]*100))
```
