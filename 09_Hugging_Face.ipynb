```python
from transformers import pipeline

classifier = pipeline("zero-shot-classification")
classifier(
    "This is a course about the Transformers library",
    candidate_labels=["education", "politics", "business"],
)
```

```python
model_name ='MoritzLaurer/multilingual-MiniLMv2-L6-mnli-xnli' # @param ["poom-sci/WangchanBERTa-finetuned-sentiment","MoritzLaurer/multilingual-MiniLMv2-L6-mnli-xnli","DAMO-NLP-SG/zero-shot-classify-SSTuning-XLM-R","Thaweewat/wangchanberta-hyperopt-sentiment-01"] {allow-input: true}
classifier = pipeline("zero-shot-classification",model=model_name)
text1 = "เราไม่ได้ทำ"
text2= "สินค้าชิ้นนี้ราคาเท่าไร"
text3="เรารักคุณ"
text4="ทำไมชีวิตต้องมาเจออะไรแบบนี้ แย่มาก ๆ"
candidate_labels = ["ประโยคคำถาม", "ประโยคบอกเล่า", "ประโยคปฏิเสธ"] # ประเภทที่เราต้องการให้แยกประเภท
classifier([text1,text2,text3,text4], candidate_labels)
```

```python
from transformers import pipeline
model_name ='Nopphakorn/mt5-small-thaisum-512-title' # @param ["Nopphakorn/mt5-small-thaisum-512-title","csebuetnlp/mT5_multilingual_XLSum"] {allow-input: true}
summarizer = pipeline("summarization",model=model_name)

text="สาขาวิชาวิทยาการคอมพิวเตอร์ เน้นการพัฒนากำลังคนให้พร้อมเข้าสู่ยุคเศรษฐกิจและสังคมดิจิทัล โดยพัฒนาบุคลากรให้มีความรู้และทักษะด้านเทคโนโลยีดิจิทัลสมัยใหม่ที่เหมาะสมต่อการดำเนินชีวิตและการประกอบอาชีพในยุคดิจิทัล รวมถึงพัฒนาทักษะที่สอดคล้องกับความต่องการในตลาดแรงงานและทักษะที่จำเป็นต่อการดำรงชีวิตในศตวรรษที่ 21 อันสอดคล้องกับยุทธศาสตร์ชาติและแผนพัฒนาเศรษฐกิจของประเทศในยุคประเทศไทย 4.0 รวมถึงพันธกิจของคณะวิทยาศาสตร์และเทคโนโลยี มหาวิทยาลัยราชภัฏสงขลาในปัจจุบัน"

summarizer(text)
```

```python
from transformers import pipeline

model_name='classla/xlm-roberta-base-multilingual-text-genre-classifier'# @param ["classla/xlm-roberta-base-multilingual-text-genre-classifier","poom-sci/WangchanBERTa-finetuned-sentiment","papluca/xlm-roberta-base-language-detection"]
classifier = pipeline("text-classification", model=model_name)

text1 = "วิทยาการคอมพิวเตอร์เป็นสาขาวิชาหนึ่ง"
text2 = "น่าเบื่อมากเลย เรียนอะไรก็ไม่รู้ เรียนไปทำไมเนี้ย"
text3 = "เป็นการเรียนที่มีความสุขมากๆเลย ชอบเรียนวิชาของวิทยาการคอมพิวเตอร์ทุกวิชาเลย สนุก ตื่นเต้น เร้าใจ ชอบมากกกกกกก"

classifier(
    [text1,text2,text3]
)
```
